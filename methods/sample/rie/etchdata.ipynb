```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
etchdata = np.genfromtxt('etchdata', skip_header=1, delimiter=',')
start = 94000
stop = 180030
time = etchdata[start:stop,0]
laser = etchdata[start:stop,1]
time = (time - time[0])/60
laser = laser/100
```

```python
%matplotlib inline

xlabel = 'time (min)'
ylabel = 'laser intensity (arb.u.)'

plt.plot(time, laser)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.savefig('etchdata.pgf')
```
